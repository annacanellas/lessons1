BONUS: Put all the previously mentioned data transformations into a function.

```python
def customeranalysis(file1):
    file1['effective to date'] = pd.to_datetime(file1['effective to date'], format= '%m/%d/%y')
    file1['month']= file1['effective to date'].dt.month
    file1[file1['effective to date'].dt.month<= 3]
    return file1
```
